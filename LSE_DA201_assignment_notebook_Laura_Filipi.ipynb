```python
#Convert the text in strings and create a new dataframe. 
tweets['text'] = tweets ['text'].astype(str)
tweets_text = tweets ['text'].apply(lambda x: x if x.strip()!= None else None)


```

```python
#Loop through the messages and create a list of values containing the # symbol

tags = []
for y in [x.split(' ') for x in tweets_text.values]:
    for z in y:
        if '#' in z:
            tags.append(z)
            
#Create a Series containing the value counts in our list 

tags = pd.Series(tags).value_counts()
            
```

```python
#Display the first 30 records 

tags.head(30)
```

```python
#Conversion of the series into a DataFrame to be used for visualisation
data = pd.DataFrame(tags).reset_index()

#rename columns

data.columns = ['word', 'count']

#Fix count datatype
data['count'] = data ['count'].astype(int)

data
```

```python
#Convert text of tweets into a list of tweets

tweets_text_values = tweets_text.values

#Convert single tweets into a list of words
tweets_text_values_words = [x.split(' ') for x in tweets_text_values]

#Empty list 
tweets_text_values_words_all = []

#Append to the list and removal of http, mentions, and selected keywords
tweets_text_values_words_all = [z for y in tweets_text_values_words for z in y if 'http' 
                                not in z and not z.startswith (('@','#')) and z.lower() not in sw_list]

#Series that counts the keywords identified
                                                                                                                       
tweets_text_values_words_all = pd.Series(tweets_text_values_words_all).value_counts() 
```
